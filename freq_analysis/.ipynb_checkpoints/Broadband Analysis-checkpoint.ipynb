```python
def build_signal(E, padfactor = 0, n_interpol = 1, dt=1)

    t = np.linspace(0, len(E),)
    t = np.linspace(0, len(t), len(t)*n_interpol)
    E = np.interp(t_interpol, t, E)
    N = len(E_interpol)
    pad_points = pad_factor * N
    i_max = np.argmax(E)
    i_min = np.argmin(E)
    i_mid = (i_max + i_min)//2
    
    left_pad = 0
    right_pad = 0
    if i_mid < N/2:
        if pad_points <= (N - 2*i_mid):
            left_pad += pad_points
            pad_points = 0
        else:
            pad_points -= (N - 2*i_mid)
            left_pad += (N - 2*i_max)
    elif i_max > N/2:
        if pad_points <= (2*i_mid - N):
            right_pad += pad_points
            pad_points = 0
        else:
            pad_points -= (2*i_mid - N)
            right_pad += (2*i_mid - N)
    right_pad += pad_points//2
    left_pad += pad_points//2
    
    E = np.concatenate((np.zeros(left_pad), E, np.zeros(right_pad)))
    t = np.linspace(0, len(E)*dt, len(E))
    
    return t, E

def get_fourier(t, E):
    spacing = t[1]-t[0]
    E_f = rfft(E)
    freq = rfftfreq(len(E), d = spacing)
    return freq, E_f

def get_broadband(freq,E_f):
    max_i = np.argmax(np.abs(E_f))
    
```
