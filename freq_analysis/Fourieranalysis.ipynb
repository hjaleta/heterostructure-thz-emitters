```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin
from scipy.fft import fft, fftfreq
# Number of sample points
```

```python
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
xf = fftfreq(N, T)[:N//2]

plt.figure()
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()

plt.figure()
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
t0 = np.arange(0,2,0.01)
t1 = np.arange(0,4,0.01)
t2 = np.arange(0,6,0.01)

freq = 1
w = 2*np.pi*freq

f0 = np.zeros(t0.shape[0])
f1 = np.sin(w*t0)
f01 = np.concatenate((f0, f1))
f010 = np.concatenate((f0, f1, f0))

fourier1 = np.abs(np.fft.rfft(f1))
fourier01 = np.abs(np.fft.rfft(f01))
fourier010 = np.abs(np.fft.rfft(f010))

freq1 = np.fft.rfftfreq(t0.shape[-1], d = 0.01)
freq01 = np.fft.rfftfreq(t1.shape[-1], d = 0.01)
freq010 = np.fft.rfftfreq(t2.shape[-1], d = 0.01)
```
